```python
import  pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import matplotlib
from pathlib import Path
plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
path = Path.cwd()
path = str(path)+'/GFMS DB/'
path
```

```python
FMSB07_자재출고 = pd.read_excel(path + 'FMSB07_자재출고.xlsx')
FMSB03_자재마스터 = pd.read_excel(path + 'FMSB03_자재마스터.xlsx',dtype={'class_cd':str})
FMSB01_자재계층 = pd.read_excel(path + 'FMSB01_자재계층.xlsx',dtype={'class_cd':str})
```

```python
FMSB03_자재마스터[['class_nm_1','class_nm_2','class_nm_3','class_nm_4','class_cd_1','class_cd_2','class_cd_3','class_cd_4']] = 0

hier = FMSB01_자재계층[['class_nm','class_level','class_cd','class_parent']]
hier = hier[hier['class_cd']!='ROOT']
hier = pd.DataFrame(hier)

def level4(i):
    # 레벨4넣기
    num = FMSB03_자재마스터.loc[i,'class_cd']
    idx = hier.index[hier['class_cd']==num]
    idx = idx[0]
    FMSB03_자재마스터.loc[i,'class_nm_4'] =hier.loc[idx,'class_nm']
    FMSB03_자재마스터.loc[i,'class_cd_4'] =hier.loc[idx,'class_cd']

    # 레벨3 넣기
    num = FMSB03_자재마스터.loc[i,'class_cd_4']
    idx = hier.index[hier['class_cd']==num]
    num_parents = hier.loc[idx[0],'class_parent']
    idx_parents = hier.index[hier['class_cd']==num_parents]
    FMSB03_자재마스터.loc[i,'class_nm_3'] =hier.loc[idx_parents[0],'class_nm']
    FMSB03_자재마스터.loc[i,'class_cd_3'] =hier.loc[idx_parents[0],'class_cd']

    #레벨2 넣기
    num = FMSB03_자재마스터.loc[i,'class_cd_3']
    idx = hier.index[hier['class_cd']==num]
    num_parents = hier.loc[idx[0],'class_parent']
    idx_parents = hier.index[hier['class_cd']==num_parents]
    FMSB03_자재마스터.loc[i,'class_nm_2'] =hier.loc[idx_parents[0],'class_nm']
    FMSB03_자재마스터.loc[i,'class_cd_2'] =hier.loc[idx_parents[0],'class_cd']
    #레벨1 넣기
    num = FMSB03_자재마스터.loc[i,'class_cd_2']
    idx = hier.index[hier['class_cd']==num]
    num_parents = hier.loc[idx[0],'class_parent']
    idx_parents = hier.index[hier['class_cd']==num_parents]
    FMSB03_자재마스터.loc[i,'class_nm_1'] =hier.loc[idx_parents[0],'class_nm']
    FMSB03_자재마스터.loc[i,'class_cd_1'] =hier.loc[idx_parents[0],'class_cd']

def level3(i):
    # 레벨3 넣기
    num = FMSB03_자재마스터.loc[i,'class_cd']
    idx = hier.index[hier['class_cd']==num]
    idx = idx[0]
    FMSB03_자재마스터.loc[i,'class_nm_3'] =hier.loc[idx,'class_nm']
    FMSB03_자재마스터.loc[i,'class_cd_3'] =hier.loc[idx,'class_cd']

    #레벨2 넣기
    num = FMSB03_자재마스터.loc[i,'class_cd_3']
    idx = hier.index[hier['class_cd']==num]
    num_parents = hier.loc[idx[0],'class_parent']
    idx_parents = hier.index[hier['class_cd']==num_parents]
    FMSB03_자재마스터.loc[i,'class_nm_2'] =hier.loc[idx_parents[0],'class_nm']
    FMSB03_자재마스터.loc[i,'class_cd_2'] =hier.loc[idx_parents[0],'class_cd']
    #레벨1 넣기
    num = FMSB03_자재마스터.loc[i,'class_cd_2']
    idx = hier.index[hier['class_cd']==num]
    num_parents = hier.loc[idx[0],'class_parent']
    idx_parents = hier.index[hier['class_cd']==num_parents]
    FMSB03_자재마스터.loc[i,'class_nm_1'] =hier.loc[idx_parents[0],'class_nm']
    FMSB03_자재마스터.loc[i,'class_cd_1'] =hier.loc[idx_parents[0],'class_cd']

def level2(i):
    #레벨2 넣기
    num = FMSB03_자재마스터.loc[i,'class_cd']
    idx = hier.index[hier['class_cd']==num]
    idx = idx[0]
    FMSB03_자재마스터.loc[i,'class_nm_2'] =hier.loc[idx,'class_nm']
    FMSB03_자재마스터.loc[i,'class_cd_2'] =hier.loc[idx,'class_cd']
    #레벨1 넣기
    num = FMSB03_자재마스터.loc[i,'class_cd_2']
    idx = hier.index[hier['class_cd']==num]
    num_parents = hier.loc[idx[0],'class_parent']
    idx_parents = hier.index[hier['class_cd']==num_parents]
    FMSB03_자재마스터.loc[i,'class_nm_1'] =hier.loc[idx_parents[0],'class_nm']
    FMSB03_자재마스터.loc[i,'class_cd_1'] =hier.loc[idx_parents[0],'class_cd']

def level1(i):
    #레벨1 넣기
    num = FMSB03_자재마스터.loc[i,'class_cd']
    idx = hier.index[hier['class_cd']==num]
    idx = idx[0]
    FMSB03_자재마스터.loc[i,'class_nm_1'] =hier.loc[idx,'class_nm']
    FMSB03_자재마스터.loc[i,'class_cd_1'] =hier.loc[idx,'class_cd']
```

```python
for i in range(len(FMSB03_자재마스터)):
    if len(FMSB03_자재마스터.loc[i,'class_cd']) == 7:
        print(3)
        level3(i)
    elif len(FMSB03_자재마스터.loc[i,'class_cd']) == 4:
        print(2)
        level2(i)
    elif len(FMSB03_자재마스터.loc[i,'class_cd']) == 2:
        print(1)
        level1(i)
```
