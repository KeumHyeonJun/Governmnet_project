```python
건축 = pd.concat([건축_2018,건축_2019,건축_2020,건축_2021,건축_2022],axis=0)
기계 = pd.concat([기계_2018,기계_2019,기계_2020,기계_2021,기계_2022],axis=0)
전기 = pd.concat([전기_2018,전기_2019,전기_2020,전기_2021,전기_2022],axis=0)
통신 = pd.concat([통신_2018,통신_2019,통신_2020,통신_2021,통신_2022],axis=0)
승강기 = pd.concat([승강기_2018,승강기_2019,승강기_2020,승강기_2021,승강기_2022],axis=0)

건축 = 건축[건축['분야별대분류'] == '건축']
기계 = 기계[기계['분야별대분류'] == '기계']
전기 = 전기[전기['분야별대분류'] == '전기']
통신 = 통신[통신['분야별대분류'] == '통신']
승강기 = 승강기[승강기['분야별대분류'] == '승강기']
```

```python
민원 = pd.concat([건축,기계,전기,통신,승강기],axis=0)
민원
```
